<b><h2>seaborn

Seaborn is a high-level statistical data visualization library built on top of Matplotlib. It is designed to simplify the creation of informative and aesthetically pleasing statistical graphics.Seaborn is particularly suited for creating visualizations of statistical relationships, distribution plots, heatmaps, pair plots, and more.

let us load our data set

```python
from google.colab import files
import pandas as pd
import seaborn as sns

# Upload the Excel file
uploaded = files.upload()

# Read the Excel file into a DataFrame
dere = pd.read_excel('pokemon_data.xlsx')

# View the data
dere.head()
```

LM plot is a function used to quickly plot the linear relationship betwean two variables.

```python
sns.lmplot(x="Attack", y="Defense", data=dere)
plt.show()
```

we can add additional column "Legendary" with hue to differentiate the colour values of  "True" and "False".

```python
sns.lmplot(x="Attack", y="Defense", data=dere , fit_reg =False, hue ="Legendary")
```

matplot lib plot



```python
fig= plt.figure()
a1 = fig.add_axes([0, 0, 1, 1])
x = np.arange(1,10)
a1.plot(x, np.exp(x), "blue")
a1.set_title("range of numbers")
#explicitly set x and y limit
plt.ylim(0,10000)
plt.xlim(0,10)
#explicitly set x and y labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

seaborn plot

```python
# Create a figure
fig = plt.figure()

# Create a subplot using Seaborn
ax = sns.lineplot(x=np.arange(1, 10), y=np.exp(np.arange(1, 10)), color="blue")

# Set the title
ax.set_title("Range of Numbers")

# Explicitly set x and y limits
ax.set_ylim(0, 10000)
ax.set_xlim(0, 10)

# Explicitly set x and y labels
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# Show the plot
plt.show()
```

**heatmap**
<p> A heatmap is a data visualization technique that uses colors to represent the values of a two-dimensional dataset. It is often used to display the relationships or correlations between variables in a matrix format, making patterns and trends easier to discern through color intensity. Heatmaps are particularly useful for visualizing complex data, such as correlation matrices, in a compact and informative way.

```python
# Create a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr)  # You can adjust the figure size as needed
```

in the above graph the correlation b/n is indicated with the colour code value in the right hand side . the most dark shows 0 correlation and the white which is among it self shows 1.

To show each with respective data values in the heatmap.

```python
# Create a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot= True)  # You can adjust the figure size as needed
```

<b> <h3>univariant visualisation </h3></b>
<p> Univariate visualization focuses on analyzing and visualizing one variable at a time, allowing us to understand its distribution, central tendencies, and key characteristics without considering its relationship with other variables. It's a useful technique to explore individual variables in isolation.

**A distplot** is a data visualization that displays the distribution of a single variable's values using a histogram with an optional kernel density estimate. It provides insights into the data's shape, central tendency, and spread, making it useful for initial data exploration and understanding the characteristics of a variable. Distplots are commonly used in data analysis and visualization tasks.

```python
sns.displot(dere1["Defense"], kde=True)
```

kde here shows the density distribution.

```python
sns.boxplot(dere1["Defense"])
```

most of the data are shifted to the left.

```python
sns.displot(dere1["Defense"], kde=False)
# we can plot with out kde only with the histogram like this.
```

we can also use only kde to plot as follows.

```python
sns.kdeplot(dere1["Defense"])
```

we can also **shade** kde for better visualisation.

```python
sns.kdeplot(dere1["Defense"], fill=True)# it was shade true now it is changed to fill true to shade the area.
```

<h3><b>Bivariate distribution visualisation <br>
joint plot</br></b></h3>
<p>
is a data visualization technique that simultaneously displays the relationship between two variables. It typically combines scatterplots, histograms, or density plots to show how the variables interact with each other, allowing you to explore their joint distribution and potential patterns or correlations in a single graph. it is useful for understanding the relationship between 2 variables in a dataset.


```python
dere1.head()
```

```python
sns.jointplot(x="Defense", y="Attack", data=dere1, kind="scatter")#with scatter plot
```

joint plot of histogram and scatter  at the same time defense and attack.

**confirm the correlation**

```python
dere1[["Defense", "Attack"]].corr()
```

```python
sns.jointplot(x="Defense", y="Attack", data=dere1, kind="kde") #with kde
```

```python
sns.jointplot(x="Defense", y="Attack", data=dere1, kind="kde", fill=True)#with kde and fill.
```

To create **a joint plot** with a **regression** line (kind="reg") using Seaborn's sns.jointplot().

```python
sns.jointplot(x="Defense", y="Attack", data=dere1, kind="reg")#with kind regression we get the below plot.
```

<b> Visualising more than two variables <br>
pair plot </b>
<p> it creates a grid of scatterplots for every combination of variables. It's a useful way to visualize pairwise relationships between variables.

```python
# Assuming "dere1" is your DataFrame with "Defense," "Attack," and "Hp" columns
sns.pairplot(data=dere1, vars=["Defense", "Attack", "HP"], kind = "scatter")
```

```python
# Assuming "dere1" is your DataFrame with "Defense," "Attack," and "Hp" columns and diagonal kind kde
sns.pairplot(data=dere1, vars=["Defense", "Attack", "HP"], kind = "scatter" , diag_kind= "kde")
```
